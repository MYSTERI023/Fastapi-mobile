# Encoding

```python
data[['Apple','Google','Motorola','Nothing','OnePlus','Oppo','Realme','Samsung','Vivo','Xiaomi']] = pd.get_dummies(data['brand'],dtype=int)
```

```python
data[['Apple A-Series','Dimensity','Exynos','Snapdragon','Tensor']] = pd.get_dummies(data['chipset_family'],dtype=int)
```

```python
data['Screen_Type_enc'] = data['screen_type'].apply(lambda s:1 if s=='LCD' else 2 if s=='AMOLED' else 3)
```

```python
# 3000 to 3500 - 1
# 3500 to 4000 - 2
# 4000 to 4500 -3
# 4500 to 5000 - 4
# 5000 to 5500 - 5
# 5500 to 6000 - 6
data['battery_enc'] = data['battery_mah'].apply(lambda b:1 if b<3500 else 2 if b<4000 else 3 if b<4500 else 4 if b<5000 else 5 if b<5500 else 6)
```

```python
X = data[['release_year', 'ram_gb', 'storage_gb','screen_size_inch',
          'refresh_rate_hz', 'camera_mp', 'supports_5g','Apple', 'Google','Motorola',
          'Nothing', 'OnePlus', 'Oppo', 'Realme', 'Samsung', 'Vivo','Xiaomi',
          'Apple A-Series', 'Dimensity', 'Exynos', 'Snapdragon','Tensor', 'battery_enc','Screen_Type_enc']]
y = data['price_inr']
```

```python
from sklearn.model_selection import train_test_split , GridSearchCV
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,random_state=23)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
scaler.fit(X_train)
X_train_scaled = scaler.transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
```

```python
Params = {'max_depth':[1,2,3,4,5,6,7,8],
         'min_samples_leaf':[2,5,7,8,10,15],
         'n_estimators':[5,10,15,20,30,50,100,200]}
```

```python
GRid = GridSearchCV(RandomForestRegressor(random_state=4),
                   param_grid=Params,
                   cv=5,
                   verbose=3,
                   n_jobs=-1)
GRid.fit(X,y)
```

```python
GRid.best_score_
```

```python
GRid.best_params_
```

```python
models = {'LR':LinearRegression(),
         'KNN3':KNeighborsRegressor(n_neighbors=3),
         'KNN5':KNeighborsRegressor(n_neighbors=5),
         'KNN7':KNeighborsRegressor(n_neighbors=7),
         'KNN9':KNeighborsRegressor(n_neighbors=9),
         'SVR2':SVR(kernel='poly',degree=2),
         'SVR3':SVR(kernel='poly',degree=3),
         "RF":RandomForestRegressor(max_depth=8,min_samples_leaf=2,n_estimators=200,random_state=23)}
```

```python
for name,model in models.items():
    model.fit(X_train_scaled,y_train)
    yp = model.score(X_test_scaled,y_test)
    print(name,yp)
```

```python
scaler2 = MinMaxScaler()
scaler2.fit(X)
X_scaled = scaler2.transform(X)
```

```python
final_model = LinearRegression()
final_model.fit(X_scaled,y)
final_model.score(X_scaled,y)
```

```python
import pickle
pickle.dump(final_model,open('final_MPP.pkl','wb'))
pickle.dump(scaler2,open('Scaled_value.pkl','wb'))
```
